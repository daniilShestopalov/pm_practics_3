```python
print("="*70)
print("–ú–û–î–£–õ–¨ 1: –ò–ù–ò–¶–ò–ê–õ–ò–ó–ê–¶–ò–Ø –î–ê–ù–ù–´–•")
print("="*70)

import numpy as np
import random

# –¥–∞–Ω–Ω—ã–µ –∏–∑ Git (34 –∑–∞–¥–∞—á–∏ –∑–∞ 11 –Ω–µ–¥–µ–ª—å)
historical_data = [5, 1, 0, 3, 0, 2, 0, 5, 4, 9, 5]

print("–î–ê–ù–ù–´–ï –ò–ó GIT:")
print(f"- –ü–µ—Ä–∏–æ–¥: 24.09 - 09.12 ({len(historical_data)} –Ω–µ–¥–µ–ª—å)")
print(f"- –ó–∞–¥–∞—á–∏ –ø–æ –Ω–µ–¥–µ–ª—è–º: {historical_data}")
print(f"- –í—Å–µ–≥–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ: {sum(historical_data)} –∑–∞–¥–∞—á")
print(f"- –°—Ä–µ–¥–Ω—è—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å: {sum(historical_data)/len(historical_data):.2f} –∑–∞–¥–∞—á–∏/–Ω–µ–¥–µ–ª—é")

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
print("\nüìà –î–ò–°–ö–†–ï–¢–ù–û–ï –†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–ï –ü–†–û–ò–ó–í–û–î–ò–¢–ï–õ–¨–ù–û–°–¢–ò:")

# –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —á–∞—Å—Ç–æ—Ç
distribution = {}
for tasks in historical_data:
    distribution[tasks] = distribution.get(tasks, 0) + 1

# –í—ã–≤–æ–¥–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
total_weeks = len(historical_data)
task_values = []
probabilities = []

for tasks, count in sorted(distribution.items()):
    task_values.append(tasks)
    prob = count / total_weeks
    probabilities.append(prob)
    print(f"  {tasks} –∑–∞–¥–∞—á–∞/–Ω–µ–¥–µ–ª—é: {prob:.3f} ({prob*100:.1f}%)")
```

```python
def monte_carlo_simulation(task_values, probabilities, tasks_needed=6, iterations=100000):
    """
    –í—ã–ø–æ–ª–Ω—è–µ—Ç —Å–∏–º—É–ª—è—Ü–∏—é –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Å—Ä–æ–∫–æ–≤ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞–¥–∞—á

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - task_values: –≤–æ–∑–º–æ–∂–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ (–∑–∞–¥–∞—á–∏/–Ω–µ–¥–µ–ª—é)
    - probabilities: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
    - tasks_needed: —Å–∫–æ–ª—å–∫–æ –∑–∞–¥–∞—á –Ω—É–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤ —Ñ–∞–∑–µ
    - iterations: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π —Å–∏–º—É–ª—è—Ü–∏–∏
    """
    results = []

    print(f" –¶–µ–ª—å: –≤—ã–ø–æ–ª–Ω–∏—Ç—å {tasks_needed} –∑–∞–¥–∞—á")
    print(f" –ó–∞–ø—É—Å–∫ {iterations:,} –∏—Ç–µ—Ä–∞—Ü–∏–π –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ...")

    for i in range(iterations):
        # –û–¥–Ω–∞ –∏—Ç–µ—Ä–∞—Ü–∏—è –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ
        completed = 0
        weeks = 0

        while completed < tasks_needed:
            # –°—ç–º–ø–ª–∏—Ä—É–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞ –Ω–µ–¥–µ–ª—é
            weekly_tasks = random.choices(task_values, weights=probabilities, k=1)[0]
            completed += weekly_tasks
            weeks += 1

        results.append(weeks)

        # –ü—Ä–æ–≥—Ä–µ—Å—Å
        if (i + 1) % 20000 == 0:
            print(f"  –ü—Ä–æ–≥—Ä–µ—Å—Å: {i + 1:,}/{iterations:,} –∏—Ç–µ—Ä–∞—Ü–∏–π")

    return np.array(results)
```

```python
print("\n" + "="*70)
print("–ú–û–î–£–õ–¨ 3: –ó–ê–ü–£–°–ö –°–ò–ú–£–õ–Ø–¶–ò–ò")
print("="*70)

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã 5-–π —Ñ–∞–∑—ã
phase_tasks = 6
iterations = 100000

# –ó–∞–ø—É—Å–∫–∞–µ–º —Å–∏–º—É–ª—è—Ü–∏—é
results = monte_carlo_simulation(
    task_values=task_values,
    probabilities=probabilities,
    tasks_needed=phase_tasks,
    iterations=iterations
)

print(f"\nüìä –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–∏–º—É–ª—è—Ü–∏–∏ —Å–æ–±—Ä–∞–Ω—ã:")
print(f"- –í—Å–µ–≥–æ –∏—Ç–µ—Ä–∞—Ü–∏–π: {len(results):,}")
print(f"- –£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {len(np.unique(results))}")
print(f"- –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è: {np.min(results)} –Ω–µ–¥–µ–ª—è")
print(f"- –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è: {np.max(results)} –Ω–µ–¥–µ–ª—å")
```

```python
print("\n" + "="*70)
print("–ú–û–î–£–õ–¨ 4: –ê–ù–ê–õ–ò–ó –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
print("="*70)

# –ë–∞–∑–æ–≤—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
mean_weeks = np.mean(results)
median_weeks = np.median(results)
std_weeks = np.std(results)

print(" –ë–ê–ó–û–í–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê:")
print(f"- –°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è: {mean_weeks:.2f} –Ω–µ–¥–µ–ª—å")
print(f"- –ú–µ–¥–∏–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è: {median_weeks:.1f} –Ω–µ–¥–µ–ª—å")
print(f"- –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_weeks:.2f} –Ω–µ–¥–µ–ª—å")

# 80% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
lower_80 = np.percentile(results, 10)  # 10-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å
upper_80 = np.percentile(results, 90)  # 90-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å

print(f"\n 80% –î–û–í–ï–†–ò–¢–ï–õ–¨–ù–´–ô –ò–ù–¢–ï–†–í–ê–õ:")
print(f"  –û—Ç {lower_80:.1f} –¥–æ {upper_80:.1f} –Ω–µ–¥–µ–ª—å")
print(f"  –ò–ª–∏ –æ—Ç {lower_80*7:.0f} –¥–æ {upper_80*7:.0f} –∫–∞–ª–µ–Ω–¥–∞—Ä–Ω—ã—Ö –¥–Ω–µ–π")
print(f"  –ò–ª–∏ –æ—Ç {lower_80*5:.0f} –¥–æ {upper_80*5:.0f} —Ä–∞–±–æ—á–∏—Ö –¥–Ω–µ–π")

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π
print(f"\n –†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–ï –í–ï–†–û–Ø–¢–ù–û–°–¢–ï–ô:")
print("  –ù–µ–¥–µ–ª—å | –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å | –ö—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è")
print("  " + "-"*40)

cumulative = 0
for weeks in range(1, int(np.max(results)) + 1):
    count = np.sum(results == weeks)
    if count > 0:
        prob = count / len(results) * 100
        cumulative += prob
        print(f"  {weeks:6} | {prob:9.1f}% | {cumulative:11.1f}%")
```

```python
# –ú–û–î–£–õ–¨ 5: –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –≤—ã–≤–æ–¥—ã –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏
print("\n" + "="*70)
print("–ú–û–î–£–õ–¨ 5: –ü–†–ê–ö–¢–ò–ß–ï–°–ö–ò–ï –í–´–í–û–î–´")
print("="*70)

print(f"–ü–†–û–ì–ù–û–ó –î–õ–Ø 5-–ô –§–ê–ó–´ ({phase_tasks} –∑–∞–¥–∞—á):")
print(f"\n1. –ù–ê–ò–ë–û–õ–ï–ï –í–ï–†–û–Ø–¢–ù–´–ô –°–†–û–ö:")
print(f"   {median_weeks:.1f} –Ω–µ–¥–µ–ª–∏ ({median_weeks*5:.0f} —Ä–∞–±–æ—á–∏—Ö –¥–Ω–µ–π)")

print(f"\n2. –° –£–ß–ï–¢–û–ú –†–ò–°–ö–û–í (80% —É–≤–µ—Ä–µ–Ω–Ω–æ—Å—Ç—å):")
print(f"    –û—Ç {lower_80:.1f} –¥–æ {upper_80:.1f} –Ω–µ–¥–µ–ª—å")
print(f"     –ë—É—Ñ–µ—Ä: +{(upper_80 - median_weeks):.1f} –Ω–µ–¥–µ–ª–∏")

print(f"\n3. –í–ï–†–û–Ø–¢–ù–û–°–¢–¨ –£–õ–û–ñ–ò–¢–¨–°–Ø –í –°–†–û–ö:")
for weeks in [1, 2, 3, 4]:
    prob = np.sum(results <= weeks) / len(results) * 100
    print(f"    –ó–∞ {weeks} –Ω–µ–¥–µ–ª—é: {prob:.1f}%")

print(f"\n4. –ö–û–ù–¢–†–û–õ–¨–ù–´–ï –¢–û–ß–ö–ò:")
print(f"    –ö–æ–Ω–µ—Ü –Ω–µ–¥–µ–ª–∏ 1: ‚â• {int(np.percentile(results, 30))} –∑–∞–¥–∞—á")
print(f"    –ö–æ–Ω–µ—Ü –Ω–µ–¥–µ–ª–∏ 2: ‚â• {int(np.percentile(results, 70))} –∑–∞–¥–∞—á")

print(f"\n5. –¢–†–ò–ì–ì–ï–†–´ –î–õ–Ø –ö–û–†–†–ï–ö–¢–ò–†–û–í–ö–ò:")
print(f"     –ï—Å–ª–∏ –Ω–µ–¥–µ–ª—è 1: < 2 –∑–∞–¥–∞—á ‚Üí –ø–µ—Ä–µ—Å–º–æ—Ç—Ä —Å—Ä–æ–∫–æ–≤")
print(f"     –ï—Å–ª–∏ –Ω–µ–¥–µ–ª—è 2: < 5 –∑–∞–¥–∞—á ‚Üí —É–≤–µ–ª–∏—á–µ–Ω–∏–µ –∫–æ–º–∞–Ω–¥—ã")

print("\n" + "="*70)
print(" –ò–¢–û–ì–û–í–´–ô –ü–†–û–ì–ù–û–ó:")
print("="*70)
print(f"–ù–∞ –æ—Å–Ω–æ–≤–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ Git –∏ –º–µ—Ç–æ–¥–∞ –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ:")
print(f" 5-—è —Ñ–∞–∑–∞ ({phase_tasks} –∑–∞–¥–∞—á) –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω–∞:")
print(f"   ‚Ä¢ –° –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 80% –∑–∞ {lower_80:.1f}-{upper_80:.1f} –Ω–µ–¥–µ–ª—å")
print(f"   ‚Ä¢ –ù–∞–∏–±–æ–ª–µ–µ –≤–µ—Ä–æ—è—Ç–Ω–æ –∑–∞ {median_weeks:.1f} –Ω–µ–¥–µ–ª—å")
print(f"\n –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è: –ø–ª–∞–Ω–∏—Ä—É–π—Ç–µ –Ω–∞ {upper_80:.1f} –Ω–µ–¥–µ–ª—å —Å –±—É—Ñ–µ—Ä–æ–º")
```

```python
# –ú–û–î–£–õ–¨ 6: –¢–µ–∫—Å—Ç–æ–≤–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
print("\n" + "="*70)
print("–ú–û–î–£–õ–¨ 6: –¢–ï–ö–°–¢–û–í–ê–Ø –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø")
print("="*70)

# –ü—Ä–æ—Å—Ç–∞—è —Ç–µ–∫—Å—Ç–æ–≤–∞—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
print("\n –ì–ò–°–¢–û–ì–†–ê–ú–ú–ê –†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–Ø (—Ç–µ–∫—Å—Ç–æ–≤–∞—è):")
print("–ù–µ–¥–µ–ª–∏ | –ß–∞—Å—Ç–æ—Ç–∞")
print("-" * 30)

for weeks in range(1, int(np.max(results)) + 1):
    count = np.sum(results == weeks)
    if count > 0:
        bar_length = int(count / len(results) * 50)
        bar = "‚ñà" * bar_length
        percent = count / len(results) * 100
        print(f"{weeks:6} | {bar} {percent:5.1f}% ({count:,})")

# –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞
print(f"\n –°–í–û–î–ù–ê–Ø –¢–ê–ë–õ–ò–¶–ê:")
print("-" * 50)
print(f"{'–ü–æ–∫–∞–∑–∞—Ç–µ–ª—å':<25} | {'–ó–Ω–∞—á–µ–Ω–∏–µ':<20}")
print("-" * 50)
print(f"{'–ò—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ':<25} | {sum(historical_data)} –∑–∞–¥–∞—á –∑–∞ {len(historical_data)} –Ω–µ–¥.")
print(f"{'–°—Ä–µ–¥–Ω—è—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å':<25} | {sum(historical_data)/len(historical_data):.2f} –∑–∞–¥–∞—á–∏/–Ω–µ–¥–µ–ª—é")
print(f"{'–ó–∞–¥–∞—á–∏ 5-–π —Ñ–∞–∑—ã':<25} | {phase_tasks} –∑–∞–¥–∞—á")
print(f"{'–ò—Ç–µ—Ä–∞—Ü–∏–π –ú–æ–Ω—Ç–µ-–ö–∞—Ä–ª–æ':<25} | {iterations:,}")
print(f"{'80% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª':<25} | {lower_80:.1f} - {upper_80:.1f} –Ω–µ–¥–µ–ª—å")
print(f"{'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∑–∞ 2 –Ω–µ–¥–µ–ª–∏':<25} | {np.sum(results <= 2)/len(results)*100:.1f}%")
print(f"{'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∑–∞ 3 –Ω–µ–¥–µ–ª–∏':<25} | {np.sum(results <= 3)/len(results)*100:.1f}%")

print("\n –í—Å–µ –º–æ–¥—É–ª–∏ —É—Å–ø–µ—à–Ω–æ –≤—ã–ø–æ–ª–Ω–µ–Ω—ã!")
```
